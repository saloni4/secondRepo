```python
import nltk
import re
#from collections import defaultdict
from nltk.probability import FreqDist
t= open("C:/Users/singh/Downloads/filteredtext.txt",encoding="utf-8").read()
unique = t.split()
fdist = FreqDist(unique)
#print(fdist.most_common(100))
fdist.hapaxes()

#sorted(w for w in set(unique) if fdist[w] <2)
for r in sorted(w for w in set(unique) if fdist[w] <2):
   
    appendFile = open('uniquewordstext.txt','a')
    appendFile.write(" "+r)
    appendFile.close()
```
