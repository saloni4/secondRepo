```python
import nltk
from nltk.tokenize import word_tokenize
```

```python
response = urllib.request.urlopen("https://en.wikipedia.org/wiki/Electric_battery")
html = response.read()
print(html)
```

```python
soup = BeautifulSoup(html,"html5lib")
text = soup.get_text(strip=True)
```

```python
tokens = [t for t in word_tokenize(text)]
```

```python
print(tokens)
```

```python
import re
words = re.findall(r'(\b[A-Za-z][a-z]{0,9}\b)',text)
#for tokens in words:
    #tokens = tokens.lower()
tokens = [w.lower() for w in words]

print(tokens)
```

```python
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
for word in tokens:
    wordd=lemmatizer.lemmatize(word)
    appendFile = open('lemmetext.txt','a')
    appendFile.write(" "+wordd)
    appendFile.close()
    print(wordd)
    
```
